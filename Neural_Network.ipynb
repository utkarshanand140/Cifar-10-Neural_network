```python
def forward_prop(x,wh,wo,bh,bo):
    zh = np.dot(x,wh) + bh                                           ## x--> 10000x1024
                                                                     ## y --> 10000x10
##  10000x64 = 10000x1024 X 1024x64                                  ## wh --> 1024x64
    H = sigmoid(zh)                                                  ## H --> 10000x64
##. 10000x64                                                         ## wo --> 64x10
                                                                     ## yhat --> 10000x10
    zo = np.dot(H,wo) + bo
##  10000x10    =  10000x64 X 64x10

    yhat = softmax(zo)
##  10000x10
    cache = {
        "zh":zh,
        "H":H,
        "zo":zo
    }
    return yhat , cache
```
